```python
from sklearn.base import BaseEstimator , TransformerMixin
```

```python
class EndofDist_Imp(BaseEstimator , TransformerMixin):
    def __init__(self , tail = "right"):
        self.tail = tail
    def fit(self , X , y = None):
        if self.tail == "right":
            self.fill_values = X.quantile(0.75)+1.5*(X.quantile(0.75)-X.quantile(0.25))
        else:
            self.fill_values = X.quantile(0.25)-1.5*(X.quantile(0.75)-X.quantile(0.25))
        return self
    def transform(self , X):
        X_filled = X.fillna(self.fill_values)
        return X_filled
```

```python
class RandomImputer(BaseEstimator , TransformerMixin):
    def __init__(self , random_state = 42):
        self.random_state = random_state
    def fit(self , X , y = None):
        X = pd.DataFrame(X)
        self.values_ = {
            col: X[col].dropna().values
            for col in X.columns
        }
        return self
    def transform(self , X):
        X = pd.DataFrame(X.copy())
        rng = np.random.default_rng(self.random_state)
        for col in X.columns:
            mask = X[col].isna()
            if mask.any():
                X.loc[mask , col] = rng.choice(self.values_[col] , size = mask.sum() , replace = True)
        return X
```

```python
def Objective(trial):
    Num_Imp = trial.suggest_categorical("Imp" , ["mean" , "median" , "eod" , "RI" , "KNN" , "MICE"])
    Cat_Imp = trial.suggest_categorical("Cat_imp" , ["ARB" , "mode" , "RI"])
    Tf = trial.suggest_categorical("Tf" , ['LT' ,  'PT_Yjohn'])
    Sca =trial.suggest_categorical("Sca" , ['Scaler' , 'RScaler' , 'MMscaler' , 'MNorm' , 'MAS'])
    if Num_Imp == "mean":
        Num_Imp = SimpleImputer(strategy = "mean")
    elif Num_Imp == "median":
        Num_Imp = SimpleImputer(strategy = "median")
    elif Num_Imp == "eod":
        Num_Imp = EndofDist_Imp()
    elif Num_Imp == "RI":
        Num_Imp = RandomImputer()
    elif Num_Imp == "KNN":
        n_neighbors = trial.suggest_int("n_neighbors" , 1 , 10 , step = 1)
        weights = trial.suggest_categorical("weights" , ['uniform', 'distance'])
        Num_Imp = KNNImputer(n_neighbors = n_neighbors , weights = weights)
    elif Num_Imp == "MICE":
        Num_Imp = IterativeImputer(sample_posterior = True , random_state = 42)
    if Cat_Imp == "ARB":
        Val = trial.suggest_categorical("Val" , ["Male" , "Female"])
        Cat_Imp = SimpleImputer(strategy = "constant" , fill_value = Val)
    elif Cat_Imp == "mode":
        Cat_Imp = SimpleImputer(strategy = "most_frequent")
    elif Cat_Imp == "RI":
        Cat_Imp = RandomImputer()
    if Tf == "LT":
        Tf = FunctionTransformer(func = np.log1p)
    elif Tf == "PT_Yjohn":
        Tf = PowerTransformer()
    if Sca == "Scaler":
        Sca = StandardScaler()
    elif Sca == "RScaler":
        Sca = RobustScaler()
    elif Sca == "MMscaler":
        Sca = MinMaxScaler()
    elif Sca == "MNorm":
        Sca = FunctionTransformer(lambda X: (X-X.mean(axis = 0)/(X.max(axis = 0)-X.min(axis = 0))))
    elif Sca == "MAS":
        Sca = MaxAbsScaler()
    Imp_CT = ColumnTransformer([
    ("Imp" , Num_Imp , Null_cols) , 
    ("Imp1" , Cat_Imp , Cat_Null_cols)
    ] , remainder = 'passthrough')
    Scaler = ColumnTransformer([
    ("Rscale" , Sca , S_List)
    ] , remainder = 'passthrough')
    Tform = ColumnTransformer([
    ("Ptrans" , Tf , S_List2)
    ] , remainder = 'passthrough')
    Encod = ColumnTransformer([
    ("OHE" , OneHotEncoder(sparse_output = False) , ['remainder__remainder__Imp1__Gender of the patient'])
    ] , remainder = 'passthrough')
    Pipe = Pipeline([
    ("Imp_CT" , Imp_CT) , 
    ("Scaler" , Scaler) , 
    ("Encod" , Encod) , 
    ("Tform" , Tform) , 
    ("LR" , LogisticRegression())
    ])
    score = cross_val_score(Pipe , X , y , scoring = "accuracy" , cv = 5 , n_jobs = -1 , error_score = 0)
    return score.mean()
```

```python
def Objective(trial):
    classifier_name = trial.suggest_categorical("Classifier_name" , ["LR" , "DTC" , "RFC" , "ABC" , "GBC" , "BGC" , "SVC" , "KNC" , "GNB" , "CBC" , "LGBC"])
    if classifier_name == "LR":
        model = LogisticRegression()
    elif classifier_name == "DTC":
        model = DecisionTreeClassifier()
    elif classifier_name == "RFC":
        model = RandomForestClassifier()
    elif classifier_name == "ABC":
        model = AdaBoostClassifier()
    elif classifier_name == "GBC":
        model = GradientBoostingClassifier()
    elif classifier_name == "BGC":
        model = BaggingClassifier()
    elif classifier_name == "SVC":
        model = SVC()
    elif classifier_name == "KNC":
        model = KNeighborsClassifier()
    elif classifier_name == "GNB":
        model = GaussianNB()
    elif classifier_name == "CBC":
        model = CatBoostClassifier()
    elif classifier_name == "LGBC":
        model = LGBMClassifier()
    Pipe = Pipeline([
    ("Imp_CT" , Imp_CT) , 
    ("Scaler" , Scaler) , 
    ("Encod" , Encod) ,
    ("Tform" , Tform) ,  
    ("model" , model)
    ])
    score = cross_val_score(Pipe , X , y , scoring = "precision" , cv = 5 , n_jobs = -1)
    return score.mean()
```

```python
study = optuna.create_study(direction = "maximize" , sampler = optuna.samplers.TPESampler())
study.optimize(Objective , n_trials = 50)
```

```python
import joblib
from fastapi import FastAPI
from pydantic import BaseModel
import sklearn
import pandas as pd 
import shap
app = FastAPI()
model = joblib.load(r"C:\FrostByte Project\Liver disease prediction model(enhanced+calibrated(sigmoid)).joblib")
rf_pipeline = model.calibrated_classifiers_[0].estimator
class InputData(BaseModel):
    Age:int
    Gender:str
    Total_Bilirubin:float
    Direct_Bilirubin:float
    Alkphos_Alkaline_Phosphotase:float
    Sgpt_Alamine_Aminotransferase:float
    Sgot_Aspartate_Aminotransferase:float
    Total_Protiens:float
    ALB_Albumin:float
    AG_Ratio_Albumin_Globulin_Ratio:float
def Dframe_Convertor(df , Transformer_name):
    return pd.DataFrame(df , columns = rf_pipeline.named_steps[Transformer_name].get_feature_names_out())
def Shap_explainations(df):
    df = rf_pipeline.named_steps["Imp_CT"].transform(df)
    df = Dframe_Convertor(df , "Imp_CT")
    df = rf_pipeline.named_steps["Scaler"].transform(df)
    df = Dframe_Convertor(df , "Scaler")
    df = rf_pipeline.named_steps["Encod"].transform(df)
    df = Dframe_Convertor(df , "Encod")
    df = rf_pipeline.named_steps["Tform"].transform(df)
    df = Dframe_Convertor(df , "Tform")
    explainer = shap.TreeExplainer(rf_pipeline.named_steps["LR"])
    shap_values = explainer(df).values[0 , : , 1]
    Index = rf_pipeline.named_steps['Tform'].get_feature_names_out()
    Clean_Index = []
    for items in Index:
        Clean_Index.append(items.split("__")[-1])
    impt = pd.Series(shap_values , index = Clean_Index)
    top_ind = impt.abs().sort_values(ascending = False).head(3)
    return top_ind
@app.post("/predict")
def Predictor(data: InputData):
    df = {
    "Age of the patient":data.Age, "Gender of the patient": data.Gender, 
    'Total Bilirubin':data.Total_Bilirubin,
       'Direct Bilirubin':data.Direct_Bilirubin, '\xa0Alkphos Alkaline Phosphotase':data.Alkphos_Alkaline_Phosphotase,
       '\xa0Sgpt Alamine Aminotransferase':data.Sgpt_Alamine_Aminotransferase, 'Sgot Aspartate Aminotransferase':data.Sgot_Aspartate_Aminotransferase ,
       'Total Protiens':data.Total_Protiens, '\xa0ALB Albumin':data.ALB_Albumin ,
       'A/G Ratio Albumin and Globulin Ratio': data.AG_Ratio_Albumin_Globulin_Ratio
    }
    df = pd.DataFrame([df])
    Exps = Shap_explainations(df)
    h_level = model.predict(df)
    response = model.predict_proba(df)[0][1]*100
    return {"Features responsible for this output": Exps , "Chances of Liver disease": f"{round(response , 2)}%" , "Risk Level":["Low" if h_level == 1 else "High"][0]}
```

```python
Data = InputData
Data.Age = 50
Data.Age = 50
Data.Gender = "Male"
Data.Total_Bilirubin = 2.3
Data.Direct_Bilirubin = 0.9
Data.Alkphos_Alkaline_Phosphotase = 1.9
Data.Sgpt_Alamine_Aminotransferase = 1.3
Data.Sgot_Aspartate_Aminotransferase = 1.87
Data.Total_Protiens = 120
Data.ALB_Albumin = 13
Data.AG_Ratio_Albumin_Globulin_Ratio = 0.98
Ans = Predictor(Data)
```
